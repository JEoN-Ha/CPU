```python
%%time
cpuData_morph_list_KR = []
cpuData_morph_cleaned_list_KR = []
for i in range(len(nn)):
    cpuData_morph_list_KR.append(nn.iloc[i][1:].tolist())
for i in range(len(cpuData_morph_list_KR)):
    cpuData_morph_cleaned_list_KR.append([x for x in cpuData_morph_list_KR[i] if str(x) != 'nan'])
```

```python
nn_kr_list = []
for kk in  cpuData_morph_cleaned_list_KR:
    for k in kk:
        nn_kr_list.append(k)
```

```python
nn_kr_list = list(set(nn_kr_list))
```

```python
wb = openpyxl.Workbook()
sheet1 = wb.active
sheet1.title = "NN_KR_List"
for i in range(len(nn_kr_list)):
    sheet1.cell(row=i+1, column=1).value = nn_kr_list[i]
wb.save('NN_KR_List.xlsx')
```
